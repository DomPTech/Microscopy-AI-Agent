```python
import os
import sys

sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), "..")))

from atomonous.config import settings

# Uncomment the line below if you want to specify the local model's download path.
# settings.hf_cache_dir = "/lustre/isaac24/scratch/dpelaia/hf_cache/"
```

```python
from atomonous.agent.core import Agent

# Initialize the agent from a local model.
# If you have it downloaded already, it will load much faster.
# You can also use Agent.from_api_key(...) depending on your setup.
agent = Agent.from_model_id(
    model_id="Qwen/Qwen2.5-14B-Instruct"
)
```

```python
# Connect to the MCP client (by default, this connects settings.mcp_url over https)
agent.connect_mcp_client(
    server_parameters={
        "url": "http://localhost:8000/mcp",
        "transport": "streamable-http"
    }
)

# Print the loaded tools
print("Available Agent Tools:")
for tool in agent.tools:
    print(f"- {tool.name}")
```

```python
# Ask the agent something
query = "Take a new image."
# You do not need to print the result (though you can)
# Because smolagents automatically streams the run
response = agent.chat(query)
```
