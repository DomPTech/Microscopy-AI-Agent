````python
import keyring
from atomonous.agent.core import Agent

api_key = keyring.get_password(
    "openrouter",
    "utkarsh"
)
````

````python
agent = Agent.from_api_key(
model_id="openrouter/google/gemma-3-27b-it",
api_base="https://openrouter.ai/api/v1",
api_key=api_key
)
````

````python
# Connect to the MCP client (by default, this connects settings.mcp_url over https)
agent.connect_mcp_client(
    server_parameters={
        "url": "http://localhost:8000/mcp",
        "transport": "streamable-http"
    }
)

# Print the loaded tools
print("Available Agent Tools:")
for tool in agent.tools:
    print(f"- {tool.name}")
````

````python
# Ask the agent something
query = "Take a new image use get_scanned_image()"
# You do not need to print the result (though you can)
# Because smolagents automatically streams the run
response = agent.chat(query)
````
